#### Defining my own Generator Function


```python
def myrangefunc(n):
    i = 0
    while i < n:
        yield i 
        i += 1
m = myrangefunc(10)
print(m)
```

```python
def myfunction(n):
    while n > 0:
        yield n ** 2
        n = n -1

for item in myfunction(4):
    print(item)
```

```python
def day():
    days = ['Sunday', 'Monday', 'Tuesday','Wednsday', 'Thursday', 'Friday','Saturday']
    i = 0 
    while True:
        yield days[i]
        i = (i+1) % 7
```

```python
m = day()
print(next(m))
print(next(m))
print(next(m))
print(next(m))
```

#### Global Variables vs Local Variables

```python
g = 10
def func():
    global g
    g = 12
    a = 1
    print(f'inside the function showing the global and local variables \ng = {g}\na = {a}')

print(f'global variable before calling the function is {g}')
func()
print(f'global variable after calling the function is {g}')
```

```python
print('Local Variables: ',locals())
print('Global Variables :',globals())
```

#### Recursive Functions


```python
def factorial(number):
    if number > 0:
        return number * factorial(number-1)
    else:
        return 1
```

```python
number = 5
print(f'Factorial of {number} is {factorial(number)}')
```

#### Nested Functions


```python
def outerfunction():
    def InnerFunction():
        print('Inner Function Calling')
    print('Outer Function Calling')
    InnerFunction()

outerfunction()
```

```python
def total_area(length, breadth, height):
    def area(dim1, dim2):
        return dim1 * dim2
    return 2 * (area(length,breadth) + area(length,height) + area(breadth,height))

print(total_area(10,5,3))
```

#### Function as Parameter

```python
def add(x,y):
    return x + y
def subtract(x,y):
    return x - y
def arithmatic(funct, x, y):
    return funct(x,y)
funct = add
print(f'The result is {arithmatic(funct,10,5)}')
funct = subtract
print(f'The result is {arithmatic(funct,10,5)}')
```

### Clouser Function
    Nested Function
    Returning Function
    Inner Function Access Outer Variables

```python
def outer_function():
    msg = 'Welcome'
    def inner_function():
        print('+' * 10)
        print(msg)
        print('-' * 10)
    return inner_function
result = outer_function()
result()
```

```python
def get_counter():
    count = 0
    def counter():
        nonlocal count # To be able to modify the higher order function variable, Have to use "nonlocal"
        count = count+1
        return count
    return counter

result = get_counter()
print(result(),result(),result(),result())
```

#### Decorative Function
Clouser Function + Function as parameter

```python
# Decorator Function
def outer_function(f):
    def inner_function():
        print('+'*10)
        f()
        print('*'*10)
    return inner_function

@outer_function
def func():
    print("Decorative Function ....")

func()
```

```python
# Decorator Function
def outer_function(f):
    def inner_function(x, y):
        print('+'*10)
        f(x,y)
        print('*'*10)
    return inner_function

@outer_function
def func(x,y):
    print(x+y)

func(10,5)
```

#### Lambda Functions

```python
def double(x):
    return x*2
print('The result of double function',double(2))
result = lambda x: x*2
print('The result of lambda function ',result(2))
```

```python
sum = lambda x,y:x+y
print('The result of lambda function is ', sum(1,3))
print('Callign lambda function for 3 + 4 is ', (lambda x,y:x+y)(3,4))
```

```python
lst = [1,2,3,4,5,6,7,8,9]
l2 = list(map(lambda x:-x,lst))
print(l2)
```

```python
lst = [1,2,3,4,5,6,7,8,9]
k  = lambda x: x if x%2==0 else -x
f = map(k,lst) # return object
print(list(f))

```
